```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
from keras.models import load_model
```

```python
new_model = load_model('cat_dog_100epochs.h5')
```

```python
import keras
import tensorflow as tf
import matplotlib.pyplot as plt

dog_file = '2.jpg'
dog_image = keras.utils.load_img(dog_file, target_size=(150, 150))
plt.imshow(dog_image)
```

```python
dog_image = keras.utils.img_to_array(dog_image)
```

```python
dog_image.shape
```

```python
import numpy as np
dog_image = np.expand_dims(dog_image, axis=0)
```

```python
dog_image.shape
```

```python
dog_image = dog_image/255
```

```python
new_model.predict(dog_image)
```

```python
predictions = (new_model.predict(dog_image) > 0.5).astype("int32")
```

```python
if new_model.predict(dog_image) > 0.5:
    print('This is a dog image!')
else:
    print('This is a cat image, not a dog!')
```
